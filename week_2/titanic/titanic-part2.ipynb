Let's aggregate data into groups and 
- get the number of entries from each class (i.e. pclass column in csv file)
- get number of entries from each town
- get general summary of each class
- get the number of entries from each class of only survivors
- plot women, men, children according to class

```python
# your code here
# Number of entries from each class
entries_per_class = titanic['pclass'].value_counts()
print("Number of entries from each class:\n", entries_per_class)

# Number of entries from each town
entries_per_town = titanic['embark_town'].value_counts()
print("\nNumber of entries from each town:\n", entries_per_town)

# General summary of each class
summary_per_class = titanic.groupby('pclass').describe().T
print("\nGeneral summary of each class:\n", summary_per_class)

# Number of entries from each class of only survivors
survivors_per_class = titanic[titanic['survived'] == 1]['pclass'].value_counts()
print("\nNumber of entries from each class of only survivors:\n", survivors_per_class)

# Plot women, men, children according to class
titanic.groupby(['who', 'class']).size().unstack().plot(kind='bar')
plt.title('Number of Women, Men, and Children')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

plot a bar graph with the number of children, women and men for each class of ticket

```python
# your code here
titanic.groupby(['who', 'class']).size().unstack().plot(kind='bar', figsize=(5,5), title="Class distribution of men, women and children on board", xlabel="class", ylabel="nb")
```

- plot a histogram that shows the number of people who survived with age on the  x-axis
- plot a histogram that shows the number of people in each class with age on the x-axis
- put the two plots beside each other horizontally

```python
# your code here
titanic[titanic['survived'] == 1]['age'].dropna().plot(kind='hist', bins=20, edgecolor='black')
plt.title('age of survivors')
plt.xlabel('age')
plt.ylabel('nb of survivors')
plt.show()

titanic['age'].dropna().plot(kind='hist', bins=20, edgecolor='black')
plt.title('class of survivors')
plt.xlabel('class')
plt.ylabel('nb of survivors')
plt.show()
```
